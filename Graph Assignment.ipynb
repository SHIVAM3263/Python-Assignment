```python
import pandas as pd
import matplotlib.pyplot as plt 
df1 = pd.read_excel('data.xlsx', header=None, skipfooter=1,sheet_name='Graph1')
# Extracting data for pie chart
title = df1.iloc[0, 0]
df1.columns = df1.iloc[1, :]  # Set the second row as column names
df1 = df1.iloc[2:8, :]  # Skip the last row and the row that is now the column names
c=["blue","red","yellow","green","darkorange","skyblue"]
plt.figure(figsize=(8, 8))
plt.pie(df1['Frequency'], labels=[f"{label}\n({percent:0.1%})" for label, percent in zip(df1['Survey Response'], df1['Percentage'])],autopct=lambda p: '{:.0f}'.format(p * sum(df1['Frequency']) / 100),startangle=90,radius=1,counterclock=False,labeldistance=1.1,rotatelabels=True,colors=c)
plt.title(title,loc='center', pad=150)
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.show()
```

```python
df_bar = pd.read_excel('data.xlsx', sheet_name='Graph2')

# Plotting the bar graph
plt.figure(figsize=(10, 6))
bar_colors=['#038cfc']*16+['red','maroon','brown']
border_colors=['red','orange','brown','none']*4+['none']*3
bars= plt.bar(df_bar['Date'], df_bar['Value'], color=bar_colors, edgecolor=border_colors)
plt.bar_label(bars, fmt='%d', label_type='edge', fontsize=8)
plt.title('Germany Birth Rate Decline')
plt.xlabel('Date')
plt.ylabel('Value')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```
